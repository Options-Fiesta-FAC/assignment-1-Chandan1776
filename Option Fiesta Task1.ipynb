```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def buy_call_payoff(spot_prices, strike_price, premium):
    return np.maximum(0, spot_prices - strike_price) - premium

def sell_call_payoff(spot_prices, strike_price, premium):
    return premium - np.maximum(0, spot_prices - strike_price)

def buy_put_payoff(spot_prices, strike_price, premium):
    return np.maximum(0, strike_price - spot_prices) - premium

def sell_put_payoff(spot_prices, strike_price, premium):
    return premium - np.maximum(0, strike_price - spot_prices)
```

```python
# Strike Price (K) and Premium (P)
K = float(input("Enter Strike Price (K): "))
P = float(input("Enter Premium (P): "))
```

```python

# Ask for spot price range
lower_bound = float(input("Enter lower bound of Spot Price range: "))
upper_bound = float(input("Enter upper bound of Spot Price range: "))

# Generate spot prices array based on user input
spot_prices = np.arange(lower_bound, upper_bound + 1)
```

```python
# Payoff calculations using defined functions
buy_call = buy_call_payoff(spot_prices, K, P)
sell_call = sell_call_payoff(spot_prices, K, P)
buy_put = buy_put_payoff(spot_prices, K, P)
sell_put = sell_put_payoff(spot_prices, K, P)
```

```python
# Setup subplots
plt.figure(figsize=(12, 8))

# Buy Call
plt.subplot(2, 2, 1)
plt.plot(spot_prices, buy_call, label='Buy Call', color='green')
plt.axhline(0, color='black', linestyle='--')
plt.title('Buying a Call Option')
plt.xlabel('Spot Price at Expiry')
plt.ylabel('Profit / Loss')
plt.legend()
plt.grid(True)

# Sell Call
plt.subplot(2, 2, 2)
plt.plot(spot_prices, sell_call, label='Sell Call', color='red')
plt.axhline(0, color='black', linestyle='--')
plt.title('Selling a Call Option')
plt.xlabel('Spot Price at Expiry')
plt.ylabel('Profit / Loss')
plt.legend()
plt.grid(True)

# Buy Put
plt.subplot(2, 2, 3)
plt.plot(spot_prices, buy_put, label='Buy Put', color='blue')
plt.axhline(0, color='black', linestyle='--')
plt.title('Buying a Put Option')
plt.xlabel('Spot Price at Expiry')
plt.ylabel('Profit / Loss')
plt.legend()
plt.grid(True)

# Sell Put
plt.subplot(2, 2, 4)
plt.plot(spot_prices, sell_put, label='Sell Put', color='orange')
plt.axhline(0, color='black', linestyle='--')
plt.title('Selling a Put Option')
plt.xlabel('Spot Price at Expiry')
plt.ylabel('Profit / Loss')
plt.legend()
plt.grid(True)

# Layout Adjustment
plt.tight_layout()
plt.suptitle('Vanilla Options Payoff Diagrams', fontsize=16, y=1.02)
plt.show()
```

```python
## Observations from Vanilla Options Payoff Diagrams

""" Buying a Call Option
- Profit occurs when the spot price at expiry is above the strike price
- The upside is unlimited as the call buyer gains when the asset price rises
- Maximum loss = Premium paid
- Breakeven point = Strike Price + Premium Paid


Selling a Call Option
- The seller profits if the spot price stays below the strike price
- Maximum profit = Premium received, achieved only if the spot price is below the strike price
- Loss is unlimited if the spot price rises significantly. This is a high risk strategy in bullish markets
- Breakeven point = Strike Price + Premium Paid

Buying a Put Option
- Profit occurs when the spot price falls below the strike price
- The downside gain is limited to Strike Price - Spot Price
- Maximum loss = Premium paid
- Breakeven point = Strike Price - Premium

Selling a Put Option
- The seller profits if the spot price stays above the strike price
- Maximum profit = Premium received
- Losses increase as the spot price drops
- Commonly used by investors who are bullish or want to acquire the stock at a lower price
- Breakeven point = Strike Price - Premium"""
```
